```python
import os
import sys
import random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, confusion_matrix,
    precision_score, recall_score, classification_report
)
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.gridspec import GridSpec
import seaborn as sns
import scipy.ndimage as ndimage
import pydicom
from PIL import Image
from concurrent.futures import ThreadPoolExecutor

SEED = 1337
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

device = "mps" if hasattr(torch.backends, "mps") and torch.backends.mps.is_available() else "cpu"
print("device:", device)
```

```python
HEALTHY_ROOT  = r"/Users/davidhernandezreyes/Downloads/manifest-1773463712896"
CANCER_ROOT   = r"/Users/davidhernandezreyes/Downloads/manifest-1646429317311"
IMG_SIZE      = (64, 64)
MAX_PER_CLASS = 3000
WL, WW        = 60, 400
class_names   = ["healthy", "cancer"]

def find_dicoms(root, max_files=None):
    out = []
    for r, _, files in os.walk(root):
        for f in files:
            if f.lower().endswith(".dcm"):
                out.append(os.path.join(r, f))
                if max_files and len(out) >= max_files:
                    return out
    return out

print("Scanning folders...")
healthy_paths = find_dicoms(HEALTHY_ROOT, MAX_PER_CLASS)
cancer_paths  = find_dicoms(CANCER_ROOT,  MAX_PER_CLASS)
print(f"  Healthy : {len(healthy_paths)} files")
print(f"  Cancer  : {len(cancer_paths)} files")
```

```python
def load_dicom(path):
    try:
        ds        = pydicom.dcmread(path, force=True)
        arr       = ds.pixel_array.astype(np.float32)
        slope     = float(getattr(ds, "RescaleSlope",     1))
        intercept = float(getattr(ds, "RescaleIntercept", 0))
        arr       = arr * slope + intercept
        lo        = WL - WW / 2
        hi        = WL + WW / 2
        arr       = np.clip(arr, lo, hi)
        arr       = (arr - lo) / (WW + 1e-7)
        im        = Image.fromarray((arr * 255).astype(np.uint8))
        im        = im.resize((IMG_SIZE[1], IMG_SIZE[0]), Image.BILINEAR)
        arr       = np.array(im, dtype=np.float32) / 255.0
        return arr[None, ...]
    except Exception:
        return None

def build_arrays(paths, label, name):
    X_list, y_list = [], []
    with ThreadPoolExecutor(max_workers=4) as ex:
        results = list(ex.map(load_dicom, paths))
    for arr in results:
        if arr is not None:
            X_list.append(arr)
            y_list.append(label)
    X = np.stack(X_list).astype(np.float32)
    y = np.array(y_list, dtype=np.int64)
    print(f"  {name}: {X.shape[0]} images loaded")
    return X, y

print("Loading images...")
X_h, y_h = build_arrays(healthy_paths, 0, "Healthy")
X_c, y_c = build_arrays(cancer_paths,  1, "Cancer")

X_all = np.concatenate([X_h, X_c])
y_all = np.concatenate([y_h, y_c])
print(f"\n  Total   : {len(X_all)} images")
print(f"  Healthy : {(y_all==0).sum()}")
print(f"  Cancer  : {(y_all==1).sum()}")
```

```python
idx = np.arange(len(X_all))
idx_tr, idx_tmp = train_test_split(
    idx, test_size=0.30, random_state=SEED, stratify=y_all
)
idx_val, idx_te = train_test_split(
    idx_tmp, test_size=0.50, random_state=SEED, stratify=y_all[idx_tmp]
)

X_train, y_train = X_all[idx_tr],  y_all[idx_tr]
X_val,   y_val   = X_all[idx_val], y_all[idx_val]
X_test,  y_test  = X_all[idx_te],  y_all[idx_te]

print(f"  Train : {len(X_train)}")
print(f"  Val   : {len(X_val)}")
print(f"  Test  : {len(X_test)}")

BATCH_SIZE   = 32
train_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train)),
    batch_size=BATCH_SIZE, shuffle=True
)
val_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val)),
    batch_size=BATCH_SIZE, shuffle=False
)
```

```python
# ── MODEL 2: DeepCNN ─────────────────────────────────────────────────
# This is the SECOND model. It differs from Model 1 (MultiCNN) in:
#   1. Extra conv layer (4 vs 3)
#   2. BatchNorm after every conv layer
#   3. Higher dropout (0.4 vs 0.25)
#   4. More channels in final layer (128 vs 64)
# These changes are justified statistically in the hypothesis test cell.

class DeepCNN(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        self.c1  = nn.Conv2d(1,   16,  3, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.c2  = nn.Conv2d(16,  32,  3, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.c3  = nn.Conv2d(32,  64,  3, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        self.c4  = nn.Conv2d(64,  128, 3, padding=1)
        self.bn4 = nn.BatchNorm2d(128)
        self.drop = nn.Dropout(0.5)
        self.drop2 = nn.Dropout(p=0.3)
        self.fc   = nn.Linear(128, num_classes)

    def forward(self, x):
        x = F.relu(self.bn1(self.c1(x))); x = F.max_pool2d(x, 2)
        x = self.drop2(x) 
        x = F.relu(self.bn2(self.c2(x))); x = F.max_pool2d(x, 2)
        x = F.relu(self.bn3(self.c3(x)))
        x = F.relu(self.bn4(self.c4(x)))
        x = x.mean(dim=(2, 3))
        x = self.drop(x)
        return self.fc(x)

model = DeepCNN(num_classes=2).to(device)
total_params = sum(p.numel() for p in model.parameters())
print("Model 2 — DeepCNN")
print(f"  Conv layers   : 4  (Model 1 had 3)")
print(f"  BatchNorm     : yes  (Model 1 had none)")
print(f"  Dropout       : 0.4  (Model 1 had 0.25)")
print(f"  Output size   : 128  (Model 1 had 64)")
print(f"  Total params  : {total_params:,}")
```

```python
EPOCHS = 100
LR     = 1e-3

opt  = torch.optim.Adam(model.parameters(), lr=LR)
crit = nn.CrossEntropyLoss()
hist = {"train_acc": [], "val_acc": [], "train_loss": [], "val_loss": []}

print(f"Training DeepCNN for {EPOCHS} epochs...")
print("=" * 55)

for epoch in range(1, EPOCHS + 1):
    model.train()
    tr_loss, tr_ok, tr_n = 0.0, 0, 0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        opt.zero_grad()
        logits = model(xb)
        loss   = crit(logits, yb)
        loss.backward()
        opt.step()
        tr_loss += loss.item() * len(yb)
        tr_ok   += (logits.argmax(1) == yb).sum().item()
        tr_n    += len(yb)

    model.eval()
    va_loss, va_ok, va_n = 0.0, 0, 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            logits = model(xb)
            loss   = crit(logits, yb)
            va_loss += loss.item() * len(yb)
            va_ok   += (logits.argmax(1) == yb).sum().item()
            va_n    += len(yb)

    hist["train_acc"].append(tr_ok / tr_n)
    hist["val_acc"].append(va_ok / va_n)
    hist["train_loss"].append(tr_loss / tr_n)
    hist["val_loss"].append(va_loss / va_n)

    if epoch % 10 == 0 or epoch == 1:
        print(f"  epoch {epoch:03d}/{EPOCHS}  "
              f"train={tr_ok/tr_n:.3f}  val={va_ok/va_n:.3f}")

best_epoch = int(np.argmax(hist["val_acc"])) + 1
print(f"\n  Best epoch : {best_epoch}  "
      f"val_acc = {hist['val_acc'][best_epoch-1]:.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

best_acc_ep  = int(np.argmax(hist["val_acc"]))
best_loss_ep = int(np.argmin(hist["val_loss"]))

ax = axes[0]
ax.plot(hist["train_acc"], label="train acc", color="steelblue")
ax.plot(hist["val_acc"],   label="val acc",   color="tomato")
ax.axvline(best_acc_ep, color="tomato", linestyle="--", alpha=0.7,
           label=f"best val acc epoch {best_acc_ep+1}")
ax.set_title("Accuracy over 100 epochs", fontweight="bold")
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.legend(fontsize=9)
ax.spines[["top", "right"]].set_visible(False)

ax = axes[1]
ax.plot(hist["train_loss"], label="train loss", color="steelblue")
ax.plot(hist["val_loss"],   label="val loss",   color="tomato")
ax.axvline(best_loss_ep, color="tomato", linestyle="--", alpha=0.7,
           label=f"best val loss epoch {best_loss_ep+1}")
ax.set_title("Loss over 100 epochs", fontweight="bold")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend(fontsize=9)
ax.spines[["top", "right"]].set_visible(False)

plt.suptitle(
    "DeepCNN learning curves — dashed line marks optimal epoch",
    fontsize=12, fontweight="bold"
)
plt.tight_layout()
plt.savefig("deepcnn_learning_curves.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
def predict(model, X):
    model.eval()
    preds = []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            xb = torch.from_numpy(X[i:i+64]).to(device)
            preds.append(model(xb).argmax(1).cpu().numpy())
    return np.concatenate(preds)

y_true = y_test.astype(np.int64)
y_pred = predict(model, X_test)

acc  = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
cm   = confusion_matrix(y_true, y_pred, labels=[0, 1])

print("=" * 55)
print("  MODEL 2 — DeepCNN TEST RESULTS")
print("=" * 55)
print(f"  Accuracy  : {acc:.4f}")
print(f"  Precision : {prec:.4f}  (macro)")
print(f"  Recall    : {rec:.4f}  (macro)")
print()
print(classification_report(y_true, y_pred,
      target_names=class_names, digits=3))

# plot confusion matrix
fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names, ax=ax)
ax.set_title("DeepCNN Confusion Matrix", fontweight="bold")
ax.set_xlabel("Predicted")
ax.set_ylabel("True")
plt.tight_layout()
plt.savefig("deepcnn_confusion_matrix.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
MODEL_INPUT_SIZE = (64, 64)
DISPLAY_SIZE     = (256, 256)

# ── CHANGE THIS TO YOUR IMAGE ──
IMAGE_PATH = r"/Users/davidhernandezreyes/Downloads/manifest-1646429317311/StageII-Colorectal-CT/StageII-Colorectal-CT-173/06-19-2007-NA-Pelvic Cavity CT RoutineEnhanced-01743/2.000000-Pelvis  5.0  B31f-03768/1-06.dcm"

def load_image(path):
    if path.lower().endswith(".dcm"):
        ds        = pydicom.dcmread(path, force=True)
        arr       = ds.pixel_array.astype(np.float32)
        slope     = float(getattr(ds, "RescaleSlope",     1))
        intercept = float(getattr(ds, "RescaleIntercept", 0))
        arr       = arr * slope + intercept
        h = arr.shape[0]
        arr = arr[int(h * 0.10):, :]
        lo, hi    = WL - WW / 2, WL + WW / 2
        arr       = np.clip(arr, lo, hi)
        arr       = (arr - lo) / (WW + 1e-7)
    else:
        im  = Image.open(path).convert("L")
        arr = np.array(im, dtype=np.float32)
        thr  = np.percentile(arr, 10)
        mask = arr > thr
        if mask.any():
            ys, xs = np.where(mask)
            y0, y1 = ys.min(), ys.max()
            x0, x1 = xs.min(), xs.max()
            pad = 5
            y0 = max(0, y0-pad); y1 = min(arr.shape[0]-1, y1+pad)
            x0 = max(0, x0-pad); x1 = min(arr.shape[1]-1, x1+pad)
            arr = arr[y0:y1+1, x0:x1+1]
        lo  = np.percentile(arr, 1)
        hi  = np.percentile(arr, 99)
        arr = np.clip(arr, lo, hi)
        arr = (arr - lo) / (hi - lo + 1e-7)

    im2  = Image.fromarray((arr * 255).astype(np.uint8)).resize(
               (DISPLAY_SIZE[1], DISPLAY_SIZE[0]), Image.BICUBIC)
    return np.array(im2, dtype=np.float32) / 255.0

def gradcam_with_temperature(model, x, temperature=3.0, target_class=None):
    model.eval()
    for p in model.parameters():
        p.requires_grad_(True)

    x_small = F.interpolate(x, size=MODEL_INPUT_SIZE,
                             mode="bilinear", align_corners=False)

    target_layer = model.c4
    store = {}
    def fwd_hook(_, __, out):  store["feat"] = out
    def bwd_hook(_, __, gout): store["grad"] = gout[0]

    h1 = target_layer.register_forward_hook(fwd_hook)
    h2 = target_layer.register_full_backward_hook(bwd_hook)

    try:
        model.zero_grad(set_to_none=True)
        logits = model(x_small)
        # ── temperature scaling softens the confidence ──
        probs = F.softmax(logits / temperature, dim=1)[0].detach().cpu().numpy()
        if target_class is None:
            target_class = int(torch.argmax(logits, dim=1).item())
        logits[0, target_class].backward()
    finally:
        h1.remove()
        h2.remove()

    weights = store["grad"].mean(dim=(2, 3), keepdim=True)
    cam     = (weights * store["feat"]).sum(dim=1, keepdim=True)
    cam     = F.relu(cam)
    cam_np  = cam[0, 0].detach().cpu().numpy()

    if cam_np.max() > 1e-8:
        cam_np = (cam_np - cam_np.min()) / (cam_np.max() - cam_np.min() + 1e-7)
        cam_np = np.power(cam_np, 0.5)
        cam_t  = torch.from_numpy(cam_np[None, None, ...])
        cam_t  = F.interpolate(cam_t, size=(512, 512),
                                mode="bicubic", align_corners=False)
        cam_np = cam_t[0, 0].numpy()
        cam_np = ndimage.gaussian_filter(cam_np, sigma=2)
        cam_np = (cam_np - cam_np.min()) / (cam_np.max() - cam_np.min() + 1e-7)
        cam_t  = torch.from_numpy(cam_np[None, None, ...])
        cam_t  = F.interpolate(cam_t, size=DISPLAY_SIZE,
                                mode="bilinear", align_corners=False)
        cam_np = cam_t[0, 0].numpy()
    else:
        cam_np = np.zeros(DISPLAY_SIZE)

    return cam_np, target_class, probs

def gradcam(model, x_input, target_class=None):
    model.eval()
    for p in model.parameters():
        p.requires_grad_(True)

    x_small = F.interpolate(x_input, size=MODEL_INPUT_SIZE,
                             mode="bilinear", align_corners=False)

    target_layer = model.c4   # DeepCNN has c4 as last conv
    store = {}
    def fwd_hook(_, __, out):  store["feat"] = out
    def bwd_hook(_, __, gout): store["grad"] = gout[0]

    h1 = target_layer.register_forward_hook(fwd_hook)
    h2 = target_layer.register_full_backward_hook(bwd_hook)

    try:
        model.zero_grad(set_to_none=True)
        logits = model(x_small)
        probs  = F.softmax(logits, dim=1)[0].detach().cpu().numpy()
        if target_class is None:
            target_class = int(torch.argmax(logits, dim=1).item())
        logits[0, target_class].backward()
    finally:
        h1.remove()
        h2.remove()

    weights = store["grad"].mean(dim=(2, 3), keepdim=True)
    cam     = (weights * store["feat"]).sum(dim=1, keepdim=True)
    cam     = F.relu(cam)
    cam_np  = cam[0, 0].detach().cpu().numpy()

    if cam_np.max() > 1e-8:
        cam_np = (cam_np - cam_np.min()) / (cam_np.max() - cam_np.min() + 1e-7)
        cam_np = np.power(cam_np, 0.5)
        cam_t  = torch.from_numpy(cam_np[None, None, ...])
        cam_t  = F.interpolate(cam_t, size=(512, 512),
                                mode="bicubic", align_corners=False)
        cam_np = cam_t[0, 0].numpy()
        cam_np = ndimage.gaussian_filter(cam_np, sigma=2)
        cam_np = (cam_np - cam_np.min()) / (cam_np.max() - cam_np.min() + 1e-7)
        cam_t  = torch.from_numpy(cam_np[None, None, ...])
        cam_t  = F.interpolate(cam_t, size=DISPLAY_SIZE,
                                mode="bilinear", align_corners=False)
        cam_np = cam_t[0, 0].numpy()
    else:
        cam_np = np.zeros(DISPLAY_SIZE)

    return cam_np, target_class, probs

if not os.path.exists(IMAGE_PATH):
    raise FileNotFoundError(f"File not found: {IMAGE_PATH}")

arr  = load_image(IMAGE_PATH)
x    = torch.from_numpy(arr[None, None, ...]).float().to(device)
heat, pred_cls, probs = gradcam_with_temperature(model, x, temperature=15.0)

label        = class_names[pred_cls]
confidence   = probs[pred_cls]
is_uncertain = (probs.max() - probs.min()) < 0.15

base_rgb = np.stack([arr] * 3, axis=-1)
heat_rgb = plt.cm.jet(heat)[..., :3]
overlay  = np.clip(0.65 * base_rgb + 0.35 * heat_rgb, 0, 1)

fig = plt.figure(figsize=(18, 5))
gs  = GridSpec(1, 4, figure=fig, wspace=0.4)

ax1 = fig.add_subplot(gs[0, 0])
ax1.imshow(arr, cmap="gray")
ax1.set_title("Input CT scan", fontsize=11, fontweight="bold")
ax1.axis("off")

ax2 = fig.add_subplot(gs[0, 1])
im2 = ax2.imshow(heat, cmap="jet", vmin=0, vmax=1)
ax2.set_title("Grad-CAM heatmap", fontsize=11, fontweight="bold")
ax2.axis("off")
plt.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04, label="activation")

ax3 = fig.add_subplot(gs[0, 2])
ax3.imshow(overlay)
ax3.set_title(f"Overlay  {label}\nConfidence: {confidence:.1%}",
              fontsize=11, fontweight="bold",
              color="gray" if is_uncertain else "darkred")
ax3.axis("off")
if is_uncertain:
    ax3.text(0.5, 0.04, "Low confidence - model uncertain",
             transform=ax3.transAxes, ha="center", fontsize=8,
             color="white",
             bbox=dict(boxstyle="round,pad=0.3", facecolor="darkorange", alpha=0.85))

ax4 = fig.add_subplot(gs[0, 3])
colors = ["tomato" if i == pred_cls else "steelblue" for i in range(len(class_names))]
bars   = ax4.barh(class_names, probs * 100, color=colors, edgecolor="white", height=0.55)
ax4.set_xlabel("Confidence (%)", fontsize=10)
ax4.set_title("All class probabilities", fontsize=11, fontweight="bold")
ax4.set_xlim(0, 105)
for bar, p in zip(bars, probs):
    ax4.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2,
             f"{p:.1%}", va="center", fontsize=9)
ax4.axvline(x=50, color="gray", linestyle="--", linewidth=0.8, alpha=0.5)
ax4.spines[["top", "right"]].set_visible(False)

plt.suptitle(
    f"Grad-CAM  |  Model 2 DeepCNN  |  Prediction: {label}  ({confidence:.1%})"
    + ("  UNCERTAIN" if is_uncertain else ""),
    fontsize=12, fontweight="bold", y=1.02,
    color="darkorange" if is_uncertain else "black"
)
plt.tight_layout()
plt.savefig("deepcnn_gradcam.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"\n{'='*50}")
print(f"  File       : {os.path.basename(IMAGE_PATH)}")
print(f"  Prediction : {label}")
print(f"  Confidence : {confidence:.1%}")
print(f"{'─'*50}")
for i, (cn, p) in enumerate(zip(class_names, probs)):
    bar    = "|" * int(p * 40)
    marker = "  < predicted" if i == pred_cls else ""
    print(f"  {cn:<10}  {p:.3f}  {bar}{marker}")
print(f"{'='*50}")
```

```python
# ── NEW SECTION: Model 1 (MultiCNN) vs Model 2 (DeepCNN) comparison ──────────

from sklearn.metrics import precision_score, recall_score
from scipy.stats import norm

# ── 1. Load DeepCNN and get predictions ──────────────────────────────────────
class DeepCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.c1   = nn.Conv2d(1, 32, 3, padding=1)
        self.bn1  = nn.BatchNorm2d(32)
        self.c2   = nn.Conv2d(32, 64, 3, padding=1)
        self.bn2  = nn.BatchNorm2d(64)
        self.c3   = nn.Conv2d(64, 128, 3, padding=1)
        self.bn3  = nn.BatchNorm2d(128)
        self.c4   = nn.Conv2d(128, 128, 3, padding=1)
        self.bn4  = nn.BatchNorm2d(128)
        self.drop = nn.Dropout(p=0.5)
        self.drop2= nn.Dropout(p=0.3)
        self.fc   = nn.Linear(128, 2)

    def forward(self, x):
        x = F.relu(self.bn1(self.c1(x))); x = F.max_pool2d(x, 2)
        x = self.drop2(x)
        x = F.relu(self.bn2(self.c2(x))); x = F.max_pool2d(x, 2)
        x = F.relu(self.bn3(self.c3(x))); x = F.max_pool2d(x, 2)
        x = F.relu(self.bn4(self.c4(x)))
        x = F.adaptive_avg_pool2d(x, 1)
        x = x.view(x.size(0), -1)
        x = self.drop(x)
        return self.fc(x)

# Load saved DeepCNN weights — adjust path to where you saved it
deep_model = DeepCNN().to(device)
deep_model.load_state_dict(torch.load("deepcnn_model.pth", map_location=device))
deep_model.eval()

# ── 2. Build a binary test set from MultiCNN's test set ──────────────────────
# DeepCNN is binary (healthy=0, cancer=1)
# MultiCNN has 4 classes (0=healthy, 1-3=cancer types)
# We convert: MultiCNN label 0 → 0 (healthy), 1/2/3 → 1 (cancer)
y_binary = (y_test > 0).astype(np.int64)

# Predict with DeepCNN
def predict_binary(model, X):
    model.eval()
    preds = []
    with torch.no_grad():
        for i in range(0, len(X), 64):
            xb = torch.from_numpy(X[i:i+64]).to(device)
            preds.append(model(xb).argmax(1).cpu().numpy())
    return np.concatenate(preds)

# MultiCNN binary predictions (collapse cancer classes)
y_pred_multi_binary = (y_pred_v2 > 0).astype(np.int64)  # V2 is your best MultiCNN
y_pred_deep         = predict_binary(deep_model, X_test)

binary_names = ["healthy", "cancer"]

# ── 3. Metrics for both models ────────────────────────────────────────────────
acc_multi = accuracy_score(y_binary, y_pred_multi_binary)
acc_deep  = accuracy_score(y_binary, y_pred_deep)

prec_multi = precision_score(y_binary, y_pred_multi_binary, average="weighted")
prec_deep  = precision_score(y_binary, y_pred_deep,         average="weighted")

rec_multi  = recall_score(y_binary, y_pred_multi_binary, average="weighted")
rec_deep   = recall_score(y_binary, y_pred_deep,         average="weighted")

print("=" * 55)
print(f"{'Metric':<20} {'MultiCNN (M1)':>15} {'DeepCNN (M2)':>15}")
print("=" * 55)
print(f"{'Accuracy':<20} {acc_multi:>15.4f} {acc_deep:>15.4f}")
print(f"{'Precision (w)':<20} {prec_multi:>15.4f} {prec_deep:>15.4f}")
print(f"{'Recall (w)':<20} {rec_multi:>15.4f} {rec_deep:>15.4f}")
print("=" * 55)

# ── 4. Confusion matrices side by side ───────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

for ax, preds, title in zip(
    axes,
    [y_pred_multi_binary, y_pred_deep],
    ["Model 1 — MultiCNN (binary collapse)", "Model 2 — DeepCNN"]
):
    cm = confusion_matrix(y_binary, preds, normalize="true")
    sns.heatmap(cm, annot=True, fmt=".2f", cmap="Blues",
                xticklabels=binary_names, yticklabels=binary_names,
                ax=ax, vmin=0, vmax=1, cbar=False)
    ax.set_title(title, fontsize=10)
    ax.set_xlabel("Predicted"); ax.set_ylabel("True")

plt.suptitle("Model 1 vs Model 2 — Normalised Confusion Matrices",
             fontsize=11, y=1.02)
plt.tight_layout()
plt.savefig("model1_vs_model2_cm.png", dpi=150, bbox_inches="tight")
plt.show()

# ── 5. McNemar's test M1 vs M2 ───────────────────────────────────────────────
c1 = (y_pred_multi_binary == y_binary)
c2 = (y_pred_deep          == y_binary)

n00 = int(( c1 &  c2).sum())
n01 = int(( c1 & ~c2).sum())
n10 = int((~c1 &  c2).sum())
n11 = int((~c1 & ~c2).sum())

print("\nMcNemar contingency table (rows=M1, cols=M2):")
print(f"              M2 correct   M2 wrong")
print(f"  M1 correct  {n00:>10d}   {n01:>8d}")
print(f"  M1 wrong    {n10:>10d}   {n11:>8d}")

discordant = n01 + n10
if discordant == 0:
    print("\nNo discordant pairs — models are identical on this test set.")
else:
    chi2_stat = (abs(n01 - n10) - 1) ** 2 / discordant
    p_value   = 1 - chi2.cdf(chi2_stat, df=1)
    print(f"\nMcNemar χ²  = {chi2_stat:.4f}")
    print(f"p-value      = {p_value:.4e}")
    if p_value < 0.05:
        print("→ REJECT H₀ — DeepCNN makes significantly different (better) errors than MultiCNN")
    else:
        print("→ FAIL TO REJECT H₀ — no significant difference detected")

# ── 6. 95% Confidence interval on accuracy difference ────────────────────────
n      = len(y_binary)
diff   = acc_deep - acc_multi
se     = np.sqrt((acc_deep*(1-acc_deep) + acc_multi*(1-acc_multi)) / n)
z      = norm.ppf(0.975)
ci_lo  = diff - z * se
ci_hi  = diff + z * se

print(f"\n95% CI on accuracy difference (DeepCNN − MultiCNN):")
print(f"  Δ = {diff:+.4f}  [{ci_lo:+.4f},  {ci_hi:+.4f}]")
if ci_lo > 0:
    print("  → CI entirely above 0: DeepCNN is significantly more accurate")
elif ci_hi < 0:
    print("  → CI entirely below 0: MultiCNN is significantly more accurate")
else:
    print("  → CI crosses 0: difference is not statistically significant")
```
