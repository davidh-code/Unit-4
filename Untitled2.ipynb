```python
import os
import pydicom
import numpy as np

PATH_HEALTHY = r"/Users/davidhernandezreyes/Downloads/TCIA_CT_COLONOGRAPHY_06-22-2015"
PATH_CANCER  = r"/Users/davidhernandezreyes/Downloads/manifest-1646429317311"

def index_dicoms(root_dir, label, max_files=2000):
    items = []
    if not os.path.exists(root_dir):
        return items

    for root, _, files in os.walk(root_dir):
        for name in files:
            if not name.lower().endswith(".dcm"):
                continue
            items.append((os.path.join(root, name), int(label)))
            if len(items) >= max_files:
                return items
    return items

def header_ok(path):
    try:
        ds = pydicom.dcmread(path, stop_before_pixels=True, force=True)
        # Basic sanity checks (doesn't touch pixel decoder)
        if not hasattr(ds, "SOPInstanceUID"):
            return False
        # Most image DICOMs have these; relax/remove if your data differs
        if not hasattr(ds, "Rows") or not hasattr(ds, "Columns"):
            return False
        return True
    except Exception:
        return False

print("Indexing paths...")
all_items = index_dicoms(PATH_HEALTHY, 0) + index_dicoms(PATH_CANCER, 1)
np.random.shuffle(all_items)

print("Validating headers...")
good, bad = [], []
for path, label in all_items:
    (good if header_ok(path) else bad).append((path, label))

print(f"Total: {len(all_items)}  Good headers: {len(good)}  Bad: {len(bad)}")
if bad[:5]:
    print("Examples bad:", [p for p, _ in bad[:5]])

# Use these for the next stage (tf.data / preprocessing)
rutas = [p for p, _ in good]
etiquetas = [l for _, l in good]
```

```python
import os
import random
import pandas as pd
from collections import Counter

SEED = 1337
random.seed(SEED)

# Cancer (3 clases por Series Description)
CANCER_ROOT = "/Users/davidhernandezreyes/Downloads/manifest-1646429317311"
CANCER_CSV  = os.path.join(CANCER_ROOT, "metadata.csv")

# Healthy (1 clase)
HEALTHY_ROOT = "/Users/davidhernandezreyes/Downloads/manifest-1773463712896"
HEALTHY_CSV  = os.path.join(HEALTHY_ROOT, "metadata.csv")

# Limite por clase para que no explote RAM/tiempo (ajusta si quieres)
MAX_PER_CLASS = 6000

cancer_series_classes = [
    "Abdomen  5.0  B30f",
    "Pelvis  5.0  B31f",
    "Venous Phase  5.0  B30f",
]

class_names = ["healthy"] + cancer_series_classes
class_to_idx = {name: i for i, name in enumerate(class_names)}
print("Classes (idx -> name):", list(enumerate(class_names)))

def to_abs_dir(manifest_root: str, file_location: str) -> str:
    s = str(file_location).strip().replace("\\", os.sep)
    if s.startswith("." + os.sep):
        s = s[2:]
    return os.path.normpath(os.path.join(manifest_root, s))

def list_dicoms(root_dir: str):
    out = []
    if not os.path.exists(root_dir):
        return out
    for r, _, files in os.walk(root_dir):
        for f in files:
            if f.lower().endswith(".dcm"):
                out.append(os.path.join(r, f))
    return out

# 1) Build datos for cancer types (3 classes)
df_c = pd.read_csv(CANCER_CSV)
df_c = df_c[df_c["Series Description"].notna()].copy()
df_c = df_c[df_c["Series Description"].isin(cancer_series_classes)].copy()

datos_by_label = {i: [] for i in range(len(class_names))}

for _, row in df_c.iterrows():
    label = class_to_idx[row["Series Description"]]
    dcm_dir = to_abs_dir(CANCER_ROOT, row["File Location"])
    datos_by_label[label].extend(list_dicoms(dcm_dir))

# 2) Build datos for healthy (1 class) from its manifest (all its DICOMs -> "healthy")
df_h = pd.read_csv(HEALTHY_CSV)
# no need for Series Description; just take all File Location folders
for _, row in df_h.iterrows():
    dcm_dir = to_abs_dir(HEALTHY_ROOT, row["File Location"])
    datos_by_label[class_to_idx["healthy"]].extend(list_dicoms(dcm_dir))

# 3) Sample up to MAX_PER_CLASS per class (to keep it manageable)
datos = []
for label, paths in datos_by_label.items():
    paths = list(dict.fromkeys(paths))  # de-dupe, keep order
    random.shuffle(paths)
    paths = paths[:MAX_PER_CLASS]
    datos.extend([(p, label) for p in paths])

random.shuffle(datos)

counts = Counter([class_names[y] for _, y in datos])
print("Total labeled DICOM images (after cap):", len(datos))
print("Counts per class:")
for name in class_names:
    print("-", name, counts.get(name, 0))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from concurrent.futures import ProcessPoolExecutor
import sys, importlib

sys.path.append("/Users/davidhernandezreyes/Documents/New project")
import dicom_mp_loader
importlib.reload(dicom_mp_loader)

SEED = 1337
IMG_SIZE = (64, 64)
WORKERS = 2

paths = [p for p, y in datos]
labels = np.array([y for p, y in datos], dtype=np.int64)

X_tr_p, X_tmp_p, y_tr, y_tmp = train_test_split(
    paths, labels, test_size=0.30, random_state=SEED, stratify=labels
)
X_va_p, X_te_p, y_va, y_te = train_test_split(
    X_tmp_p, y_tmp, test_size=0.50, random_state=SEED, stratify=y_tmp
)

train_files = list(zip(X_tr_p, y_tr.tolist()))
val_files   = list(zip(X_va_p, y_va.tolist()))
test_files  = list(zip(X_te_p, y_te.tolist()))

print("Splits:", "train", len(train_files), "val", len(val_files), "test", len(test_files))
print("Class balance train:", {i:int((y_tr==i).sum()) for i in range(len(class_names))})

def build_split_arrays(files, name):
    jobs = [(p, y, IMG_SIZE[0], IMG_SIZE[1]) for (p, y) in files]
    X_list, y_list = [], []
    with ProcessPoolExecutor(max_workers=WORKERS) as ex:
        for arr, label, ok, path in ex.map(dicom_mp_loader.load_one_from_tuple, jobs, chunksize=64):
            if ok:
                X_list.append(arr)
                y_list.append(label)
    X = np.stack(X_list, axis=0).astype(np.float32)  # (N,1,H,W)
    y = np.array(y_list, dtype=np.int64)
    print(f"{name}: X={X.shape} y={y.shape}")
    return X, y

X_train, y_train = build_split_arrays(train_files, "TRAIN")
X_val,   y_val   = build_split_arrays(val_files,   "VAL")
X_test,  y_test  = build_split_arrays(test_files,  "TEST")

print("Classes (index -> name):", list(enumerate(class_names)))
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

SEED = 1337
EPOCHS = 120
BATCH_SIZE = 16
LR = 1e-3

torch.manual_seed(SEED)
np.random.seed(SEED)

device = "cpu"
if hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
    device = "mps"
print("device:", device)

train_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train)),
    batch_size=BATCH_SIZE, shuffle=True
)
val_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val)),
    batch_size=BATCH_SIZE, shuffle=False
)

class MultiCNN(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        self.c1 = nn.Conv2d(1, 16, 3, padding=1)
        self.c2 = nn.Conv2d(16, 32, 3, padding=1)
        self.c3 = nn.Conv2d(32, 64, 3, padding=1)
        self.drop = nn.Dropout(0.25)
        self.fc = nn.Linear(64, num_classes)

    def forward(self, x):
        x = F.relu(self.c1(x)); x = F.max_pool2d(x, 2)
        x = F.relu(self.c2(x)); x = F.max_pool2d(x, 2)
        x = F.relu(self.c3(x))
        x = x.mean(dim=(2, 3))          # global avg pool
        x = self.drop(x)
        return self.fc(x)               # logits (N,C)

model = MultiCNN(num_classes=len(class_names)).to(device)
opt = torch.optim.Adam(model.parameters(), lr=LR)
crit = nn.CrossEntropyLoss()

hist = {"train_acc": [], "val_acc": [], "train_loss": [], "val_loss": []}

for epoch in range(1, EPOCHS + 1):
    model.train()
    tr_loss, tr_ok, tr_n = 0.0, 0, 0
    for xb, yb in train_loader:
        xb = xb.to(device)
        yb = yb.to(device)

        opt.zero_grad()
        logits = model(xb)
        loss = crit(logits, yb)
        loss.backward()
        opt.step()

        tr_loss += loss.item() * yb.numel()
        pred = logits.argmax(dim=1)
        tr_ok += int((pred == yb).sum().item())
        tr_n += int(yb.numel())

    model.eval()
    va_loss, va_ok, va_n = 0.0, 0, 0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb = xb.to(device)
            yb = yb.to(device)
            logits = model(xb)
            loss = crit(logits, yb)

            va_loss += loss.item() * yb.numel()
            pred = logits.argmax(dim=1)
            va_ok += int((pred == yb).sum().item())
            va_n += int(yb.numel())

    tr_acc = tr_ok / tr_n
    va_acc = va_ok / va_n
    hist["train_acc"].append(tr_acc)
    hist["val_acc"].append(va_acc)
    hist["train_loss"].append(tr_loss / tr_n)
    hist["val_loss"].append(va_loss / va_n)

    if epoch % 10 == 0 or epoch == 1:
        print(f"epoch {epoch:03d}/{EPOCHS} train_acc={tr_acc:.3f} val_acc={va_acc:.3f}")

best_epoch = int(np.argmax(hist["val_acc"])) + 1
print("Best epoch by val_acc:", best_epoch, "val_acc:", hist["val_acc"][best_epoch - 1])

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(hist["train_acc"], label="train_acc")
plt.plot(hist["val_acc"], label="val_acc")
plt.title("Accuracy")
plt.xlabel("epoch"); plt.ylabel("acc"); plt.legend()

plt.subplot(1,2,2)
plt.plot(hist["train_loss"], label="train_loss")
plt.plot(hist["val_loss"], label="val_loss")
plt.title("Loss")
plt.xlabel("epoch"); plt.ylabel("loss"); plt.legend()

plt.tight_layout()
plt.show()

# ---- TEST evaluation ----
model.eval()
preds = []
with torch.no_grad():
    for i in range(0, len(X_test), 32):
        xb = torch.from_numpy(X_test[i:i+32]).to(device)
        logits = model(xb)
        preds.append(logits.argmax(dim=1).cpu().numpy())

y_pred = np.concatenate(preds, axis=0).astype(np.int64)
y_true = y_test.astype(np.int64)

cm = confusion_matrix(y_true, y_pred, labels=list(range(len(class_names))))
acc = accuracy_score(y_true, y_pred)
prec_macro = precision_score(y_true, y_pred, average="macro", zero_division=0)
rec_macro = recall_score(y_true, y_pred, average="macro", zero_division=0)

print("\n== Multiclass Model (test) ==")
print("Accuracy:", acc)
print("Precision (macro):", prec_macro)
print("Recall (macro):", rec_macro)
print("Confusion matrix (rows=true, cols=pred):\n", cm)
print("Class names:", list(enumerate(class_names)))

plt.figure(figsize=(6,6))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xticks(range(len(class_names)), range(len(class_names)))
plt.yticks(range(len(class_names)), range(len(class_names)))
for (i,j), v in np.ndenumerate(cm):
    plt.text(j, i, str(v), ha="center", va="center", fontsize=8)
plt.xlabel("pred")
plt.ylabel("true")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import torch
from PIL import Image

PNG_PATH = "/Users/davidhernandezreyes/Downloads/colon_cancer.png"  # cambia esto

def preprocess_png_robust(png_path, out_size=(64, 64)):
    im = Image.open(png_path).convert("L")
    arr = np.array(im, dtype=np.float32)

    # 1) recorte simple: elimina borde negro usando umbral
    thr = np.percentile(arr, 10)  # robusto
    mask = arr > thr
    if mask.any():
        ys, xs = np.where(mask)
        y0, y1 = ys.min(), ys.max()
        x0, x1 = xs.min(), xs.max()
        # un poco de margen
        pad = 5
        y0 = max(0, y0 - pad); y1 = min(arr.shape[0]-1, y1 + pad)
        x0 = max(0, x0 - pad); x1 = min(arr.shape[1]-1, x1 + pad)
        arr = arr[y0:y1+1, x0:x1+1]

    # 2) normaliza por percentiles (evita outliers)
    lo = np.percentile(arr, 1)
    hi = np.percentile(arr, 99)
    arr = np.clip(arr, lo, hi)
    arr = (arr - lo) / (hi - lo + 1e-7)

    # 3) resize a 64x64
    im2 = Image.fromarray((arr * 255).astype(np.uint8)).resize(out_size)
    arr2 = np.array(im2, dtype=np.float32) / 255.0

    x = torch.from_numpy(arr2[None, None, ...]).float()  # (1,1,H,W)
    return x

def predict_png(png_path):
    model.eval()
    x = preprocess_png_robust(png_path).to(device)
    with torch.no_grad():
        logits = model(x)
        probs = torch.softmax(logits, dim=1)[0].cpu().numpy()
    pred = int(np.argmax(probs))
    print("file:", png_path)
    print("PRED:", pred, class_names[pred], "prob:", float(probs[pred]))
    print("All probs:")
    for i, name in enumerate(class_names):
        print(f"{i}: {name}  {probs[i]:.4f}")
    return pred, probs

pred, probs = predict_png(PNG_PATH)
```

```python
import numpy as np
import pydicom
import torch

DICOM_PATH = r"/Users/davidhernandezreyes/Downloads/manifest-1646429317311/StageII-Colorectal-CT/StageII-Colorectal-CT-229/01-24-2006-NA-Pelvic Cavity CT RoutineEnhanced-19716/2.000000-Pelvis  5.0  B31f-76297/1-41.dcm"

IMG_SIZE = (64, 64)

def resize_simple(arr, target_h=64, target_w=64):
    h, w = arr.shape[:2]
    step_h = max(1, h // target_h)
    step_w = max(1, w // target_w)
    small = arr[::step_h, ::step_w][:target_h, :target_w]
    out = np.zeros((target_h, target_w), dtype=np.float32)
    out[: small.shape[0], : small.shape[1]] = small
    return out

def preprocess_dicom(dicom_path):
    ds = pydicom.dcmread(dicom_path, force=True)
    arr = ds.pixel_array.astype(np.float32)
    mn, mx = float(arr.min()), float(arr.max())
    arr = (arr - mn) / (mx - mn + 1e-7)
    arr = resize_simple(arr, IMG_SIZE[0], IMG_SIZE[1])
    x = torch.from_numpy(arr[None, None, ...]).float()  # (1,1,H,W)
    return x

def predict_dicom_multiclass(dicom_path, topk=4):
    model.eval()
    x = preprocess_dicom(dicom_path).to(device)

    with torch.no_grad():
        logits = model(x)  # (1,C)
        probs = torch.softmax(logits, dim=1)[0].cpu().numpy()

    order = np.argsort(-probs)[:topk]
    print("file:", dicom_path)
    for i in order:
        print(f"{i}: {class_names[i]}  prob={probs[i]:.4f}")
    pred = int(order[0])
    print("PRED:", pred, class_names[pred], "prob=", float(probs[pred]))
    return pred, probs

pred, probs = predict_dicom_multiclass(DICOM_PATH)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class MultiCNN(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        self.c1 = nn.Conv2d(1, 16, 3, padding=1)
        self.c2 = nn.Conv2d(16, 32, 3, padding=1)
        self.c3 = nn.Conv2d(32, 64, 3, padding=1)
        self.drop = nn.Dropout(0.25)
        self.fc = nn.Linear(64, num_classes)

    def forward(self, x):
        x = F.relu(self.c1(x)); x = F.max_pool2d(x, 2)
        x = F.relu(self.c2(x)); x = F.max_pool2d(x, 2)
        x = F.relu(self.c3(x))
        x = x.mean(dim=(2, 3))
        x = self.drop(x)
        return self.fc(x)

device = "mps" if hasattr(torch.backends, "mps") and torch.backends.mps.is_available() else "cpu"
model = MultiCNN(num_classes=len(class_names)).to(device)

from torchinfo import summary
summary(model, input_size=(1, 1, 64, 64), col_names=("input_size","output_size","num_params"))
```
